```python
# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("SampleSuperstore.csv", delimiter=",")
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
# Find whether there are any missing values.
data.isnull().sum()
```

<p fontsize="4">There are no missing values in the dataset</p>

```python
# Check for duplicates in the dataset
data.duplicated().sum()
```

```python
# Drop the duplicates
data.drop_duplicates()
```

```python
# Find number of distinct observations 
data.nunique()
```

```python
# Remove unneccsary columns
column = ['Postal Code']
data1 = data.drop(columns=column,axis=1)
```

```python
data1.info()
```

```python
# Correlation between variables
data1.corr()
```

```python
# Covariance of columns
data1.cov()
```

```python
data1.head()
```

```python
# Find Category and sub-category with most sales
data1["Category"].value_counts().idxmax()
```

```python
data1["Sub-Category"].value_counts().idxmax()
```

## Data Visualization

```python
plt.hist(data1["Category"])
plt.show()
```

```python
plt.figure(figsize=(12, 12))
plt.title("Sub-Category")
plt.pie(data1["Sub-Category"].value_counts(), labels=data1["Sub-Category"].value_counts().index, autopct="%1.1f%%")
plt.show()
```

```python
plt.title("Category")
plt.pie(data1["Category"].value_counts(), labels=data1["Category"].value_counts().index, autopct="%1.1f%%")
plt.show()
```

```python
plt.title("Region wise Profits")
plt.pie(data1["Region"].value_counts(), labels=data1["Region"].value_counts().index, autopct="%1.1f%%")
plt.show()
```

## Profit and Loss

```python
# Check for High Profits
profit = data1[data1["Profit"]>3000]
profit
```

```python
# Check for Loss
loss = data1[data1["Profit"]<0]
loss
```

```python
plt.figure(figsize=(8,8))
sns.barplot(x=profit["Profit"], y=profit["City"], data=data1)
```

```python
data1.groupby('Category')[['Profit', 'Sales']].sum().plot.bar(color=['pink', 'red'], alpha=0.9, figsize=(8, 5))
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

## As a business manager, try to find out the weak areas where you can work to make more profit.

<li>Technology and Office Supplies have high profit
<li>Furniture have less profit

```python
x = data1.groupby('Sub-Category')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
x[:].plot.bar(color=['purple', 'blue'], figsize=(15, 8))
plt.title('Profit/Loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/Loss & Sales')
plt.show()
```

<li>Phones have high sales.</li>
<li>Chairs have high sales,but less profit when compared to phones.</li>
<li>Tables and Bookcases are facing hge losses.</li>
